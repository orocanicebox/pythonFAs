### Analyze songs by artist and views-youtube, stream-spotify.

### Create bar charts for Top 10 Artists on Youtube and Top 10 Artists on Spotify in 1 row of display.

```python
df_views = df.groupby('Artist')['Views'].sum().sort_values(ascending=False) [:10]
df_streams = df.groupby('Artist')['Stream'].sum().sort_values(ascending=False) [:10]

fig, (ax1, ax2) = mpl.subplots(1,2, figsize=(12,6))

ax1.set_title("Top 10 Artists on Youtube")
df_views.plot(kind="bar", ax=ax1)

ax2.set_title("Top 10 Artists on Spotify")
df_streams.plot(kind="bar", ax=ax2)

ax1.set_xlabel("Artists")
ax1.set_ylabel("Views")

ax2.set_xlabel("Artists")
ax2.set_ylabel("Streams")
fig.tight_layout()

mpl.show()
```

The data showed in figure 1 (left) is the top 10 artists with the most views on Youtube while the showed in figure 2 (right)is the top 10 artists with the most streams on Spotify.

Ed Sheeran has the highest number of views with estimated 1.6 billion views in total. on the other hand, he came second with the number of streams on Spotify after Post Malone.

### Determine the top 10 artists on the Youtube platform whose songs have more likes and display a bar chart with a title Number of Likes per Artist and make some remarks.

```python
df_artists_likes = df.groupby(['Artist'])[['Likes']].sum().sort_values(by=['Likes'], ascending=False).head(10)

df_artists_likes.reset_index()
fig, (ax1) = mpl.subplots(1, figsize=(14,10))

ax1.set_title("Top 10 Artists with the most Likes")
df_artists_likes .plot(kind="barh", ax=ax1)
ax1.set_xlabel("Artists")
ax1.set_ylabel("Likes")

mpl.show()
```

BTS is the artist that has the highest Likes with 140 million likes followed by BLACKPINK with a total of 130 Million likes. The next after BLACKPINK are artists with more than 50 million likes less than them.

### Explore single releases by artists and identify top performers based on likes and display a table for each artist with their Views and Likes as the other columns.


```python
df_singles = df[df.Album_type == 'single']
grouped = df_singles.groupby("Artist").agg({ "Likes": "sum","Views": "sum"})

df_artist_singles = grouped.sort_values("Likes", ascending=False).head(10)

df_artist_singles
```

in this we can see that BLACKPINK is the artist that has the highest likes and views when it comes to singles releases. they are then followed by Wiz Khalifa who came in second on both the likes and views aspect. They are followed by different artists with almost the same number of likes, however Bizzarap who came in at 8th place on video likes had the lowest views. 

```python
df_compy = df[df.Album_type == 'compilation']
grouped = df_compy.groupby("Artist").agg({ "Likes": "sum","Views": "sum"})

df_artist_compy = grouped.sort_values("Likes", ascending=False).head(10)

df_artist_compy
```

The dat shows that Shakira is the artist that has the highest amounts of likes and views for relasing compilation of musics.She is followed by Ellie goulding who has almost half of Shakira's records in terms of Likes. From the data we can see that Annie Lenox has the least amount of streams and likes from the top 10
